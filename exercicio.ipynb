Executar um teste de performance para avaliar com quais valores de k 1 a 25 o algoritmo knn tem a melhor performance

```python
#import do dataset
from sklearn.datasets import load_iris
#Carregar o dataset
iris = load_iris()
#Separando os dados em x e y
x = iris.data
y = iris.target
#Importação da função necessária para dividir o datset
from sklearn.model_selection import train_test_split
#Seperando o dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
#Importando o knn
from sklearn.neighbors import KNeighborsClassifier
#Lista para armazenar as performances para os vários k
performance = {}
#Importando a mátrica para validação dos resutlados
from sklearn import metrics
for k in range(1, 26):
    #Inicializando o knn com o k
    knn = KNeighborsClassifier(n_neighbors = k)
    #Realizando o treinamento
    knn.fit(x_train, y_train)
    #Verificando os resultados para o conjeuto de teste
    previsoes = knn.predict(x_test)
    #Avaliando o a acurácia do modelo
    acuracia = metrics.accuracy_score(y_test, previsoes)
    performance[k] = round(acuracia,4)
print(performance)
    

```

```python
import matplotlib.pyplot as plt
#Separando o dicionário por posições, em uma lista ficam as chaves e em outra os dados
plt.plot(list(performance.keys()),list(performance.values()))
plt.xlabel('Valores de k')
plt.ylabel('Performance')
plt.show()
```
