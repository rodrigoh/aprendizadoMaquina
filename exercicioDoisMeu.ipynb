##### Exercício para avaliar o impacto de cada uma das variáveis sobre o resultado

```python
# imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
#Importando o banco de dados
publi = pd.read_csv('dados/advertising.csv')
# Imprimindo as 6 primeiras linhas 
#publi.head()
# REALIZANDO A ANÁLISE COM TODOS OS DADOS
# Observações
x = publi[['TV','Radio','Newspaper']]
# Variável alvo
y = publi['Sales']
# Separando o conjunto de dados em treino e teste
#random_state=5 faz com que mesmo que seja realizado um random ele mantenha os valores a cada execução
#com ele faremos os testes com os mesmos registros, mesmo removendo uma variável
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
# Criando a instância do modelo Regressão Linear
reglin = LinearRegression()
# Rodando o modelo
reglin.fit(x_train, y_train)
# Realizando as previsões
y_prev = reglin.predict(x_test)
# Avaliando o modelo
rmseTresVariaveis = np.sqrt(metrics.mean_squared_error(y_test, y_prev))
#######################################################################
# Realizando os testes com os recursos TV e Rádio
# Observações
x = publi[['TV','Radio']]
# Variável alvo
y = publi['Sales']
# Separando o conjunto de dados em treino e teste
#random_state=5 faz com que mesmo que seja realizado um random ele mantenha os valores a cada execução
#com ele faremos os testes com os mesmos registros, mesmo removendo uma variável
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
# Criando a instância do modelo Regressão Linear
reglin = LinearRegression()
# Rodando o modelo
reglin.fit(x_train, y_train)
# Realizando as previsões
y_prev = reglin.predict(x_test)
# Avaliando o modelo
rmseTvRadio = np.sqrt(metrics.mean_squared_error(y_test, y_prev))
#######################################################################
# Realizando os testes com os recursos TV e Jornal
# Observações
x = publi[['TV','Newspaper']]
# Variável alvo
y = publi['Sales']
# Separando o conjunto de dados em treino e teste
#random_state=5 faz com que mesmo que seja realizado um random ele mantenha os valores a cada execução
#com ele faremos os testes com os mesmos registros, mesmo removendo uma variável
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
# Criando a instância do modelo Regressão Linear
reglin = LinearRegression()
# Rodando o modelo
reglin.fit(x_train, y_train)
# Realizando as previsões
y_prev = reglin.predict(x_test)
# Avaliando o modelo
rmseTvJornal = np.sqrt(metrics.mean_squared_error(y_test, y_prev))
#######################################################################
# Realizando os testes com os recursos TV e Jornal
# Observações
x = publi[['Radio','Newspaper']]
# Variável alvo
y = publi['Sales']
# Separando o conjunto de dados em treino e teste
#random_state=5 faz com que mesmo que seja realizado um random ele mantenha os valores a cada execução
#com ele faremos os testes com os mesmos registros, mesmo removendo uma variável
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
# Criando a instância do modelo Regressão Linear
reglin = LinearRegression()
# Rodando o modelo
reglin.fit(x_train, y_train)
# Realizando as previsões
y_prev = reglin.predict(x_test)
# Avaliando o modelo
rmseRadioJornal = np.sqrt(metrics.mean_squared_error(y_test, y_prev))
# Imprimindo os resultados
print('Valores de Erro Quadrático Médio da Raiz')
print('TV, Radio, Newspaper:', np.round(rmseTresVariaveis, 3))
print('TV, Radio:', np.round(rmseTvRadio, 3))
print('TV, Newspaper:', np.round(rmseTvJornal, 3))
print('Radio, Newspaper:', np.round(rmseRadioJornal, 3))
```
